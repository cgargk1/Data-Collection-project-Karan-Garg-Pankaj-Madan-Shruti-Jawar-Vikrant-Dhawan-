```python
ps = openpyxl.load_workbook('ProwessSeed.xlsx')
sheet = ps['Sheet1']
```

```python
dx=[]
for row in range(2,51809):
    # Creating URL based on CIN number and CompanyName 
    CIN = sheet['A' + str(row)].value
    CNAME = sheet['B' + str(row)].value.strip().replace(' ','-')
    # Login to Company Check
    urlC='https://www.thecompanycheck.com/company/'+ CNAME+'/'+ CIN
    values = {'username': 'shrutijhawar13@gmail.com',
          'password': 'isb2022$'}
    page = requests.get(urlC, data=values)
    #Get attribute Values
    if (page.status_code==200):
        print("Data Extracted",CNAME+'/'+ CIN)
        tree = html.fromstring(page.content)
        
        #NSE
        StockSymbolNSEText=(tree.xpath('//*[@id="About"]/div[4]/table/tbody/tr[3]/td[2]/div/div[2]/span/text()'))
        if len(StockSymbolNSEText) == 0:
            StockSymbolNSE=""
        else:
            StockSymbolNSE=StockSymbolNSEText[0].strip()
        
        #BSE
        StockSymbolBSEText=(tree.xpath('//*[@id="About"]/div[4]/table/tbody/tr[3]/td[2]/div/div/span/text()'))          
        if (len(StockSymbolBSEText) == 0):
            StockSymbolBSE=""
        else:
            StockSymbolBSE=StockSymbolBSEText[0].strip()
            
        #ROC
        RocCodeText=(tree.xpath('//*[@id="td_roc"]/text()'))          
        if (len(RocCodeText) == 0):
            RocCode=""
        else:
            RocCode=RocCodeText[0].strip()
        
        #CompanyNumber
        CompanyNoText=(tree.xpath('//*[@id="td_regnumber"]/text()'))
        if (len(CompanyNoText) == 0):
            CompanyNo=""
        else:
            CompanyNo=CompanyNoText[0].strip()
        #IncorpDate
        IncorpDateText=(tree.xpath('//*[@id="td_incorporationDate"]/text()'))
        if (len(IncorpDateText) == 0):
            IncorpDate=""
        else:
            IncorpDate=IncorpDateText[0].replace('\n',"")
        #DateofAGM
        DateOfAgmText=(tree.xpath('//*[@id="td_dateOfAgm"]/text()'))
        if (len(DateOfAgmText) == 0):
            DateOfAgm=""
        else:
            DateOfAgm=DateOfAgmText[0].replace('\n',"")          
        #listing Status           
        ListingStatusText=(tree.xpath('//*[@id="td_listingStatus"]/span/text()'))
        if (len(ListingStatusText) == 0):
            ListingStatus=""
        else:
            ListingStatus=ListingStatusText[0].strip()
        #Industry
        IndustryText=(tree.xpath('/html/body/div[5]/div[1]/div[1]/div[4]/div[2]/div[2]/div[3]/div[1]/div[4]/table/tbody/tr[15]/td[2]/text()'))
        if (len(IndustryText) == 0):
            Industry=""
        else:
            Industry=IndustryText[0].strip()
        #Segment
        SegmentText=(tree.xpath('/html/body/div[5]/div[1]/div[1]/div[4]/div[2]/div[2]/div[3]/div[1]/div[4]/table/tbody/tr[16]/td[2]/text()'))
        if (len(SegmentText) == 0):
            Segment=""
        else:
            Segment=SegmentText[0].strip()
        #Company Size
        CompanySizeText=(tree.xpath('/html/body/div[5]/div[1]/div[1]/div[4]/div[2]/div[2]/div[3]/div[1]/div[4]/table/tbody/tr[17]/td[2]/text()'))
        if (len(CompanySizeText) == 0):
            CompanySize=""
        else:
            CompanySize=CompanySizeText[0].strip()
        #Company Details
        CompanyDetailsText=(tree.xpath('//*[@id="About"]/p/span/text()'))
        if (len(CompanyDetailsText) == 0):
            CompanyDetails=""
        else:
            CompanyDetails='It'+CompanyDetailsText[1].replace('\n ',"")
        #openLoans
        OpenLoansText=(tree.xpath('//*[@id="dv_chargeFound"]/p/b/text()'))
        if (len(OpenLoansText) == 0):
            OpenLoans=0
        else:
            OpenLoans=int(OpenLoansText[0].strip())
        #TotalSecuredAmt       
        TotalSecuredAmtText=(tree.xpath('//*[@id="dv_chargeFound"]/div[1]/span/text()'))
        if (len(TotalSecuredAmtText) == 0):
            TotalSecuredAmt=0
        else:
             TotalSecuredAmt=TotalSecuredAmtText[0].replace('\n₹ ',"").replace('\n',"").replace(',',"").replace('₹ ',"")
        #Website
        WebsiteText=tree.xpath('//*[@id="dv_website"]/a/text()')
        if (len(WebsiteText) == 0):
            Website=""
        else:
            Website=WebsiteText[0]
        
            
        #Company Class
        companyClassText=(tree.xpath('//*[@id="td_companyclass"]/text()'))
        if len(companyClassText) == 0:
            companyClass=""
        else:
            companyClass=companyClassText[0].strip()
        #CompanyCategory
        companyCategoryText=(tree.xpath('//*[@id="td_companyCategory"]/text()'))
        if len(companyCategoryText) == 0:
            companyCategory=""
        else:
            companyCategory=companyCategoryText[0].strip()
        
        #CompanySubCategory
        companySubCategoryText=(tree.xpath('//*[@id="td_compantsub"]/text()'))
        if len(companySubCategoryText) == 0:
            companySubCategory=""
        else:
            companySubCategory=companySubCategoryText[0].strip()
    
        #AuthorizedCapital
        authCapText=(tree.xpath('//*[@id="td_authorisedCapital"]/text()'))
        if (len(authCapText) == 0):
            AuthorizedCapital=0
        else:
             AuthorizedCapital=authCapText[0].replace('\n₹ ',"").replace('\n',"").replace(',',"").replace('₹ ',"")
        
        #State From Location
        
        StateText=(tree.xpath('//*[@id="contact-details"]/tbody/tr[1]/td[2]/text()'))
        if len(StateText) == 0:
            State=""
        else:
            StateTemp=StateText[0]
            State=StateTemp[StateTemp.find(",")+1:].strip()
        
        dx.insert(row,[CIN,companyClass,StockSymbolNSE,StockSymbolBSE,RocCode,CompanyNo,IncorpDate,DateOfAgm,ListingStatus,Industry,Segment,CompanySize,CompanyDetails,OpenLoans,TotalSecuredAmt,Website,companyClass,companyCategory,companySubCategory,AuthorizedCapital,State])
        
    else:
        print('URL NOT FOUND',CNAME+'/'+ CIN)

df=pd.DataFrame(dx,columns=['CIN','companyClass','StockSymbolNSE','StockSymbolBSE','RocCode','CompanyNo','IncorpDate','DateOfAgm','ListingStatus','Industry','Segment','CompanySize','CompanyDetails','OpenLoans','TotalSecuredAmt','Website','companyClass','companyCategory','companySubCategory','AuthorizedCapital','State'])
df.set_index('CIN')
df
```

```python
#Saving Extracted Data to csv
df.to_csv('ExDataCC.csv')
```
